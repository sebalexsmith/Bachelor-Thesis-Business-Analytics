```python
import pandas as pd
import numpy as np
```

```python
liste20 = pd.read_excel('20.xlsx')
liste35 = pd.read_excel('35.xlsx')
liste54 = pd.read_excel('54.xlsx')
liste109 = pd.read_excel('109.xlsx')
liste156 = pd.read_excel('156.xlsx')
liste164 = pd.read_excel('164.xlsx')
liste324 = pd.read_excel('324.xlsx')
```

```python
# List of file names
file_list = [liste20 ,liste35, liste54, liste109, liste156, liste164, liste324]

def concat_files(file_list):
    # Initialize an empty list to store dataframes
    dfs = []
    
    # Loop through the file list and read each Excel file into a dataframe
    for file_name in file_list:
        dfs.append(file_name)
    
    # Concatenate all dataframes in the list along the rows (axis=0)
    concatenated_df = pd.concat(dfs, axis=0, ignore_index=True)
    
    return concatenated_df

# Call the function to concatenate the files
concatenated_df = concat_files(file_list)
concatenated_df
```

```python
concatenated_df.to_excel("combined7092-slutt.xlsx", index=False)
```
