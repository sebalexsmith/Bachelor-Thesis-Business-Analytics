```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../Felles/TrondheimHousingDataCleaned.csv', sep = ',')
blr = pd.read_csv('../Nore/boliglånsrente.csv', encoding = 'latin-1', sep = ';')
```

```python
blr
```

```python
blr['Year'] = blr['måned']
for i in range(len(blr['måned'])):
    blr['Year'][i] = blr['måned'][i][:4]
    blr['måned'][i] = blr['måned'][i][5:]
```

```python
type(blr['Year'][0])
```

```python
years = ['2021', '2022', '2023']
HLR = blr[blr['Year'].isin(years)]
HLR['Year'] = HLR['Year'].astype(int)
HLR
```

```python
HLR.rename(columns = {'måned': 'Month',
                         'Renter på nye boliglån': 'Borrowing rate%'}, inplace = True)
HLR['Month'] = HLR['Month'].astype(int)
HLR.reset_index(inplace = True, drop= 'First')
```

```python
HLR
```

```python
#Housing price index dataframe
bpi = pd.read_csv('../Nore/boligprisindeks.csv', sep = ';')
bpi
```

```python

bpi['Year'] = bpi['Kvartal']
for i in range(len(bpi['Kvartal'])):
    bpi['Year'][i] = bpi['Kvartal'][i][:4]
    bpi['Kvartal'][i] = bpi['Kvartal'][i][5]
```

```python
bpi.rename(columns = {'Kvartal': 'Quarter',
                         'Hele_landet': 'HPI Norway',
                         'Trondheim': 'HPI Trondheim'}, inplace = True)
bpi.reset_index(inplace = True, drop= 'First')
```

```python
years = ['2021', '2022', '2023']
HPI = bpi[bpi['Year'].isin(years)]
HPI['Year'] = HPI['Year'].astype(int)
```

```python
HPI
```

```python
a = pd.concat([HPI] * 3 , ignore_index=True) 
a.sort_values(by= ['Year', 'Quarter'], inplace = True)
a['Quarter'] = a['Quarter'].astype(int)
```

```python
a = a.reset_index(drop = True)
```

```python
a['Month'] = 0
```

```python
for i in range(3 * 12):
    a['Month'][i] = (i % 12) + 1
```

```python
HPI = a
HPI
```
