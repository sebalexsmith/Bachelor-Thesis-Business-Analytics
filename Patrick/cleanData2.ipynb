```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel('../Felles/TrondheimHousingData.xlsx')
```

```python
df.drop(columns = ['Hjemmelshaver'], inplace = True)
```

```python
df['Postnummer'] = df['Adresse'].str.split(', ').str[1].str[:4].astype(int)
```

```python
df = pd.get_dummies(df, columns = ['Balkong', 'Parkering'], dtype = int)
df = df.drop(['Parkering_Nei'], axis = 1)
```

```python
df['Registrert dato'] = pd.to_datetime(df['Registrert dato'][0], infer_datetime_format = True)
df['Sist solgt'] = pd.to_datetime(df['Sist solgt'][0], infer_datetime_format = True)
```

```python
numeric_columns = ['Antall rom', 'Omsetningshastighet (dager)', 'Antall soverom', 'Etasje', 'Byggeår', 'BRA', 'BTA', 'Prom']
for column in numeric_columns:
    df[column] = df[column].fillna(0)

df['Antall rom'] = df['Antall rom'].astype(int)
df['Omsetningshastighet (dager)'] = df['Omsetningshastighet (dager)'].astype(int)
df['Antall soverom'] = df['Antall soverom'].astype(int)
df['Etasje'] = df['Etasje'].astype(int)
df['Byggeår'] = df['Byggeår'].astype(int)
df['BRA'] = df['BRA'].astype(int)
df['BTA'] = df['BTA'].astype(int)
df['Prom'] = df['Prom'].astype(int)
```

```python
# Define the mapping of old names to new names
name_mapping = {
    'heimdal eiendomsmegling': 'Heimdal Eiendomsmegling',
    'proaktiv': 'Proaktiv Eiendomsmegling',
    'eiendomsmegler 1': 'EiendomsMegler 1',
    'eie': 'EIE',
    'nylander': 'Meglerhuset Nylander',
    'privatmegleren': 'Priv.Megleren',
    'krogsveen': 'Krogsveen',
    'dnb': 'DNB Eiendom',
    'obos': 'OBOS'
}

# Replace the values in the 'Megler' column based on the mapping
df['Megler'] = df['Megler'].str.lower().replace(name_mapping)

# For values not found in the mapping, set them to 'Annet'
df['Megler'] = df['Megler'].fillna('Annet')
```
