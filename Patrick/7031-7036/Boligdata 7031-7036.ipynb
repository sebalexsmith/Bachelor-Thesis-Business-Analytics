```python
import pandas as pd
import numpy as np
```

```python
liste92 = pd.read_excel('92.xlsx')
liste136 = pd.read_excel('136.xlsx')
liste173 = pd.read_excel('173.xlsx')
liste210 = pd.read_excel('210.xlsx')
liste275 = pd.read_excel('275.xlsx')
liste325 = pd.read_excel('325.xlsx')
liste406= pd.read_excel('406.xlsx')
```

```python
# List of file names
file_list = [liste92 ,liste136, liste173, liste210, liste275, liste325, liste406]

def concat_files(file_list):
    # Initialize an empty list to store dataframes
    dfs = []
    
    # Loop through the file list and read each Excel file into a dataframe
    for file_name in file_list:
        dfs.append(file_name)
    
    # Concatenate all dataframes in the list along the rows (axis=0)
    concatenated_df = pd.concat(dfs, axis=0, ignore_index=True)
    
    return concatenated_df

# Call the function to concatenate the files
concatenated_df = concat_files(file_list)
concatenated_df
```

```python
concatenated_df.to_excel("combined7031-7036.xlsx", index=False)
```
