```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("boligm_info.xlsx")
```

```python
df = df.drop("Boliger i alt", axis=1)
```

```python
df.head(1)
```

```python
df.rename(columns={'Unnamed: 1': 'Year'}, inplace=True)
df.rename(columns={'Prisindeks for brukte boliger': 'Housing Price Index Trondheim'}, inplace=True)
df.rename(columns={'Boligendring': 'Change in Housing price index'}, inplace=True)
df.rename(columns={'KPI': 'Consumer Price Index'}, inplace=True)
df.rename(columns={'Endring KPI': 'Change Consumer Price Index'}, inplace=True)
df.rename(columns={'Styringsrente': 'Policy Rate %'}, inplace=True)
df.rename(columns={'Styringsrente.1': 'Policy Rate'}, inplace=True)
df.rename(columns={'Endring Styringsrente': 'Change Policy Rate'}, inplace=True)
```

```python
df = df[4:]
df = df.reset_index(drop=True)
```

```python
df.head(1)
```

```python
# Plot the line plot
ax = df.plot(x='Year', y=['Housing Price Index Trondheim', 'Consumer Price Index'], kind='line')

ax.set_xticks(df['Year'])

# Set the labels and title
plt.xlabel('Year')
plt.ylabel('Index')
plt.title('Housing price index and Consumer price index')
plt.xticks(rotation=45)

# Show the plot
plt.grid(True, alpha = 0.2)
plt.legend()
plt.show()
```

```python
# Plot the line plot
ax = df.plot(x='Year', y=['Change in Housing price index', 'Policy Rate'], kind='line')

ax.set_xticks(df['Year'])

# Set the labels and title
plt.xlabel('Year')
plt.title('Housing Price Index and Policy Rate')
plt.xticks(rotation=45)

# Show the plot
plt.grid(True, alpha = 0.2)
plt.legend()
plt.show()
```
